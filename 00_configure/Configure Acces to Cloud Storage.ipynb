#### Configure Acces to Cloud Storage via Unity Catalog

```python
%sql
SHOW STORAGE CREDENTIALS
```

```python
%sql 
SHOW CATALOGS;
```

```python
%sql
  
CREATE EXTERNAL LOCATION IF NOT EXISTS `databricks_kackis`
URL 'abfss://demo@kackistorage.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `databricks-credential`)
COMMENT 'External location for demo container';
```

```python
%fs
ls 'abfss://demo@kackistorage.dfs.core.windows.net/'
```

```python
%sql
  
CREATE EXTERNAL LOCATION IF NOT EXISTS `databricks_kackis_formula1`
URL 'abfss://formula1@kackistorage.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `databricks-credential`)
COMMENT 'External location for formula1 container';
```

```python
%fs
ls 'abfss://formula1@kackistorage.dfs.core.windows.net/'
```

```python
%sql
CREATE CATALOG IF NOT EXISTS formula1
  MANAGED LOCATION 'abfss://formula1@kackistorage.dfs.core.windows.net/'
  COMMENT 'Main catalog for formula1 project';   
```

#### Create schema

```python
%sql
CREATE SCHEMA IF NOT EXISTS formula1.landing;
CREATE SCHEMA IF NOT EXISTS formula1.bronze
  MANAGED LOCATION 'abfss://formula1@kackistorage.dfs.core.windows.net/bronze';
CREATE SCHEMA IF NOT EXISTS formula1.silver
  MANAGED LOCATION 'abfss://formula1@kackistorage.dfs.core.windows.net/silver';
CREATE SCHEMA IF NOT EXISTS formula1.gold
  MANAGED LOCATION 'abfss://formula1@kackistorage.dfs.core.windows.net/gold';
```

```python
%sql
SHOW SCHEMAS
```

```python
%sql
SELECT current_catalog();
```

```python
%sql
USE CATALOG formula1;  
SHOW SCHEMAS
```

#### Create volume files

```python
%sql
CREATE EXTERNAL VOLUME formula1.landing.files
LOCATION 'abfss://formula1@kackistorage.dfs.core.windows.net/landing';
```
